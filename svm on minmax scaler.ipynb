```python
import pandas as pd

earthquake = pd.read_csv('Earthquake_data_dataset.csv')
earthquake.head()
```

```python
maximum_threshold = earthquake['magnitude'].quantile(0.95)

earthquake[earthquake['magnitude']>maximum_threshold]
```

```python
minimum_threshold = earthquake['magnitude'].quantile(0.05)

earthquake[earthquake['magnitude']>minimum_threshold]
```

```python
earthquake_no_outliers = earthquake[(earthquake['magnitude']> minimum_threshold) & (earthquake['magnitude']< maximum_threshold)]
earthquake_no_outliers
```

```python
x = earthquake_no_outliers.iloc [:, 1:]
y = earthquake_no_outliers.iloc [:, 0]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.svm import SVR

model = SVR()
model.fit(x_train, y_train)
```
